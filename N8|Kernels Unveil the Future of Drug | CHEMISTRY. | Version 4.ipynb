## <span style="color: #FF4500;"> LOADING THE DATASET </span>


```python
df = pd.read_csv("/kaggle/input/delaney-solubility-with-descriptors/delaney_solubility_with_descriptors.csv")
```

## <span style="color: #FF4500;"> DATA PREPARATION. </span>



## <span style="color: #FF4500;"> 1- DEFINING INDEPENDENT AND DEPENDENT VARIABLES: </span>

X = INDEPENDENT OR THE TARGET VARIABLE. <BR>
Y = DEPENDENT OR PREDICTOR VARIABLE.


```python
y = df["logS"]
y #The dependent variable.
```

```python
x = df.drop("logS",axis = 1)
x #The independent variable.
```

## <span style="color: #FF4500;"> Splitting the datasets into Training and Testing data : </span>
### In this case we will be dividing and keeping the Training dataset as 80%(0.8) and the remaining Testing data as 20%(0.2).

```python
x_train , x_test  = train_test_split(x,train_size = 0.8 , random_state = 100) 
```

```python
y_train,y_test = train_test_split(y,train_size = 0.8,random_state = 100)
```

## <span style="color: #FF4500;"> 4.) MODEL BUILDING :  </span>
<p style="padding: 10px; background-color: #000000; margin: 0; color: #FF4500; font-family: New Times Roman; font-size: 135%; text-align: center; border-radius: 5px; overflow: hidden; font-weight: 500;">

## <span style="color: #FF4500;"> 1.) LINEAR REGRESSION MODEL: </span>


#### <span style="color: #FF4500;"> Training the model</span>


```python
lr = LinearRegression()
lr.fit(x_train , y_train)
```

## <span style="color: #FF4500;"> 4.1 : Random Forest Regressor model. </span>


#### <span style="color: #FF4500;"> Training the model </span>


```python
rfr = RandomForestRegressor(max_depth = 2 , random_state = 100)
rfr.fit(x_train,y_train)
```

## <span style="color: #FF4500;"> 5.) PREDICTION. </span>
<p style="padding: 10px; background-color: #000000; margin: 0; color: #FF4500; font-family: New Times Roman; font-size: 135%; text-align: center; border-radius: 5px; overflow: hidden; font-weight: 500;">
                     STEP-5: PREDICTION.<br>
 
</p>

<p style="text-align: center;">
   

## <span style="color: #FF4500;"> 5.1 : Linear regression model prediction:</span>


Since we already know Y is a dependent variable that depends on value of the independent variable X so in form of statistics it can be represented as ;

Y =F(X)

## LR TRAINING DATA

```python
y_lr_pred_train = lr.predict(x_train)
print(y_lr_pred_train)

#Prediction for the train or the 80% of data.
```

## TESTING DATA OF LR

```python
y_lr_pred_test = lr.predict(x_test)
y_lr_pred_test
# Prediction for the test or remaining 20% of data.
```

## <span style="color: #FF4500;"> 5.2 : Random Forest Regressor model prediction: </span>


## Randomforest Training data

```python
y_rfr_train_pred = rfr.predict(x_train)
y_rfr_train_pred
```

## Randomforest Testing data :

```python
y_rfr_test_pred = rfr.predict(x_test)
y_rfr_test_pred
```

## <span style="color: #FF4500;"> 6.) EVALUATION: </span>
<p style="padding: 10px; background-color: #000000; margin: 0; color: #FF4500; font-family: New Times Roman; font-size: 135%; text-align: center; border-radius: 5px; overflow: hidden; font-weight: 500;">
                     STEP-6 : EVALUATION:<br>
  EVALUATION or CHECKING THE MODEL PERFORMANCE :
</p>

<p style="text-align: center;">
 

## <span style="color: #FF4500;"> 6.1 : Linear regression model. </span>


### <span style="color: #FF4500;"> 6.1.1  : Checking the accuracy of Y prediction for the training data </span>

We will be using , mean_squared_error and R*2 statistics here.

```python
lr_train_mse = mean_squared_error(y_train,y_lr_pred_train)
lr_train_r2 = r2_score(y_train , y_lr_pred_train)
```

### <span style="color: #FF4500;"> Checking the accuracy of Y prediction for the testing data </span>


```python
lr_test_mse = mean_squared_error(y_test,y_lr_pred_test)
lr_test_r2 = r2_score(y_test , y_lr_pred_test)
```

```python
print("The MSE values for Linear regression model of training data is  = {} and for Testing data it is = {}".format(lr_train_mse,lr_test_mse))
```

```python
print("The R*2 statistics score values for Linear regression model of training data is  = {} and for Testing data it is = {}".format(lr_train_r2,lr_test_r2))
```

### <span style="color: #FF4500;"> Creating a table for the evaluation of all models. </span>


```python
evaluation = pd.DataFrame(['Linear Regression'  , lr_train_mse, lr_test_mse,lr_train_r2,lr_test_r2]).transpose()
evaluation.columns = ['Model name','Training data MSE' , 'Testing data MSE' , 'R*2 score Training data', 'R*2 score Testing data' ]
```

## <span style="color: #FF4500;"> 6.2 : Random Forest Model. </span>


### <span style="color: #FF4500;"> 6.2.1  : Checking the accuracy of Y prediction for the training data </span>

We will be using , mean_squared_error and R*2 statistics here.

```python
rfr_train_mse = mean_squared_error(y_train,y_rfr_train_pred)
rfr_train_r2_score = r2_score(y_train,y_rfr_train_pred)
```

### <span style="color: #FF4500;"> 6.2.2  : Checking the accuracy of Y prediction for the testing data </span>

We will be using , mean_squared_error and R*2 statistics here.

```python
rfr_test_mse = mean_squared_error(y_test,y_rfr_test_pred)
rfr_test_r2_score = r2_score(y_test,y_rfr_test_pred)
```

```python
print("The MSE values for Random Forest model of training data is  = {} and for Testing data it is = {}".format(rfr_train_mse,rfr_test_mse))
```

```python
print("The  values for Random Forest model of training data is  = {} and for Testing data it is = {}".format(rfr_train_r2_score,rfr_test_r2_score))
```

```python
evaluation2 = pd.DataFrame(['Random Forest Model'  , rfr_train_mse, rfr_test_mse,rfr_train_r2_score,rfr_test_r2_score]).transpose()
evaluation2.columns = ['Model name','Training data MSE' , 'Testing data MSE' , 'R*2 score Training data', 'R*2 score Testing data' ]
evaluation2
```

## <span style="color: #FF4500;"> 6.3 : FINAL EVALUATION. </span>


```python
final_evaluation = pd.concat([evaluation , evaluation2],axis=0)
```

Reseting the index for the dataframe.

```python
final_evaluation.reset_index(drop=True)
```
