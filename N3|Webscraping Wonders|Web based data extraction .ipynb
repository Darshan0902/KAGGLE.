#### <span style="color: #FF4500;"> 2.) Libraries and modules and how to install them : 
 
 1.) beautifulsoup4 4.12.2 :
         Beautiful Soup is a library that makes it easy to scrape information from web pages. It sits atop an HTML or XML parser, providing Pythonic idioms for iterating, searching, and modifying the parse tree. It helps us basically to read the HTML files we have on any particular websites. To install this library use the syntax below : 
         
```

```   
    
 To know more about this module refer : https://pypi.org/project/beautifulsoup4/

````python
# import the beautiful soup library :
import bs4 as bs

# Import the API requests.
import requests

# Passing the API requests through a variable named HTML .
html = requests.get("https://en.wikipedia.org/wiki/List_of_S%26P_500_companies")

# Retreiving the whole page into html.text format into a variable named as "soup" to store the parsed page into soup variable.
soup = bs.BeautifulSoup(html.text,'html.parser')

# Defining a empty list with the name "tickers".
tickers = []


# Using the "find" generally returns all the table so as result I ensured that we include the only needed "table" instead of
# the whole page .
table = soup.find("table",{"class" : "wikitable sortable"})

# Taking the rows into variable rows.
rows = table.findAll('tr')[1:]

# Retrieving all the rows in table
for row in rows:
    ticker = row.findAll('td')[0].text
    tickers.append(ticker[:-1])
    # [:-1] was used to remove the "/n" we will be recieving in outputs.

    print(ticker)
````
