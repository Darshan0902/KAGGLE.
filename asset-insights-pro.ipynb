## <span style="color: #FF4500;"> I - BALANCE SHEET .<br><div>  
    
A balance sheet is a financial statement that provides a snapshot of a company's financial position at a specific point in time. It presents a summary of a company's assets, liabilities, and shareholders' equity, also known as net worth or book value.

The balance sheet follows the fundamental accounting equation:

Assets = Liabilities + Shareholders' Equity

Here's a breakdown of the components of a balance sheet:

1. **Assets**:
    Assets represent what a company owns or controls, and they are categorized into current assets and non-current (or long-term) assets. Current assets include cash, accounts receivable, inventory, and short-term investments, while non-current assets include property, plant, and equipment, investments, and intangible assets.

2. **Liabilities**:
    Liabilities represent what a company owes to external parties and are categorized into current liabilities and non-current (or long-term) liabilities. Current liabilities include accounts payable, short-term loans, and accrued expenses, while non-current liabilities include long-term debt, deferred tax liabilities, and pension obligations.

3. **Shareholders**
    Equity: Shareholders' equity represents the residual interest in the company's assets after deducting liabilities. It includes the initial capital invested by shareholders plus retained earnings, which are the accumulated profits or losses retained in the business over time.

The balance sheet provides valuable information about a company's liquidity, solvency, and overall financial health. It is often analyzed in conjunction with other financial statements, such as the income statement and cash flow statement, to gain a comprehensive understanding of a company's performance and financial position.

# <a id="0.1.1"></a>
## <span style="color: #FF4500;">II - TOTAL CURRENT ASSETS .<br><div>  
    
Total current assets, as the name suggests, represents the sum of all current assets on a company's balance sheet. Current assets are assets that are expected to be converted into cash or used up within one year or one operating cycle, whichever is longer.
    

# <a id="0.2.1"></a>
## <span style="color: #FF4500;">III - TOTAL CURRENT LIABILITIES. <br><div>  
Total current liabilities, as the name suggests, represents the sum of all current liabilities on a company's balance sheet. Current liabilities are obligations or debts that are expected to be settled within one year or one operating cycle, whichever is longer.

# <a id="0.2"></a>
## <span style="color: #FF4500;">IV -  CASH DEBT DIFFERENCE. <br><div>  
In the provided code segment, the calculation of the `cash_debt_difference` represents the difference between the cash and cash equivalents and the total debt of a company. The result can be positive or negative, and the interpretation depends on the context.

- If the `cash_debt_difference` is positive, it means that the company has more cash and cash equivalents than its total debt. This indicates a healthier financial position, as the company has sufficient liquid assets to cover its debt obligations.

- If the `cash_debt_difference` is negative, it means that the company has less cash and cash equivalents than its total debt. This suggests a relatively weaker financial position, as the company may face challenges in meeting its debt obligations with the available liquid assets.

The positive or negative value of the `cash_debt_difference` provides insights into the company's liquidity and its ability to manage debt. A positive value is generally more favorable, indicating a stronger financial standing, while a negative value may indicate potential financial strain.

# <a id="0.2.1"></a>
## <span style="color: #FF4500;"> V - GOODWILL & INTANGIBLE ASSETS : 
 <br><div> 
**Intangible Assets**: Intangible assets are assets that lack physical substance but hold value for a company. They typically arise from intellectual property, brand recognition, patents, trademarks, copyrights, customer relationships, and proprietary technology. Intangible assets are not easily measurable or tangible but can contribute significantly to a company's value and competitive advantage.

**Goodwill**: Goodwill represents the excess amount paid for an acquired business over the fair value of its identifiable assets and liabilities. It arises from factors such as the acquired company's reputation, customer loyalty, skilled workforce, and strategic location. Goodwill is considered an intangible asset and represents the value of the synergies and potential future earnings expected from the acquisition.

# <a id="1"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" > 1 -  Importing Libraries and Setting Variables .<br><div>  

```python
# Importing the requests library for making HTTP requests
import requests  

# Importing the matplotlib library for plotting
import matplotlib.pyplot as plt   

# API key for accessing the financial data
api_key = "92b75130d3daacde926058020c7591da" 

# Ticker symbol of the company for which data will be fetched
company = "GOOG"  

# Number of years for which balance sheet data will be retrieved
years = 5  
```

# <a id="2"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" >2 -  Fetching Balance Sheet Data<br><div>

```python
# Retreiveing Data or requests into a variable through Web Scraping : 
balance_sheet = requests.get(f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/{company}?periods=quarterly&limit={years}&apikey={api_key}')

# Formatting the response data as JSON format
balance_sheet = balance_sheet.json() 

# Printing the data . 
balance_sheet
```

# <a id="3"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" >3 - Calculating and Printing Current Assets.<br><div>

```python
# Extracting total current assets from the balance sheet
total_current_assets = balance_sheet[0]['totalCurrentAssets']  

# Printing the total current assets
print(f"Total current assets of {company}: {total_current_assets:,}") 
```

# <a id="4"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" >4 - Calculating and Printing Current Liabilities<br><div>

```python
# Extracting total current liabilities
total_current_liabilities = balance_sheet[0]['totalCurrentLiabilities']  

# Printing the total current liabilities
print(f"Total current liabilities of {company}: {total_current_liabilities:,}")  
```

# <a id="5"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" >5 - Calculating and Printing Cash Debt Difference.<br><div>

```python
 # Extracting the total debt
total_debt = balance_sheet[0]['totalDebt'] 

# Extracting cash and cash equivalents
cash_and_equivalents = balance_sheet[0]['cashAndCashEquivalents']  

# Calculating the cash debt difference
cash_debt_difference = cash_and_equivalents - total_debt  

# Printing the cash debt difference
print(f"Cash debt difference: {cash_debt_difference:,}")  
```


# <a id="6"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" >6 - Calculating And Printing Percentage Of Intangible Assets<br><div>

```python
# Extracting goodwill and intangible assets
goodwill_and_intangibles = balance_sheet[0]['goodwillAndIntangibleAssets']  

 # Extracting the total assets
total_assets = balance_sheet[0]['totalAssets'] 

 # Calculating the percentage of intangible assets
percentage_intangible = goodwill_and_intangibles / total_assets 

# Printing the percentage of intangible assets
print(f"Percentage intangible: {percentage_intangible * 100}%")  
```


# <a id="7"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" >7 - Defining Quarterly Assets Data

```python
# Extracting total assets for quarter 1
assets_q1 = balance_sheet[4]['totalAssets']

# Extracting total assets for quarter 2
assets_q2 = balance_sheet[3]['totalAssets'] 

# Extracting total assets for quarter 3
assets_q3 = balance_sheet[2]['totalAssets']  

# Extracting total assets for quarter 4
assets_q4 = balance_sheet[1]['totalAssets'] 
```


# <a id="8"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" >8 - Scaling Assets Data and Plotting

```python
# making a  list of this assets :
assets_data = [assets_q1 , assets_q2 , assets_q3 , assets_q4]

#Scaling the values:
assets_data = [x / 1000000000 for x in assets_data]

# Plotting a bar plot.
plt.bar([1,2,3,4] , assets_data)
plt.title("Quarterly assets data of the : {company}")
plt.xlabel("Quarters")
plt.ylabel("Total assets in BILLIONS (USA DOLLARS : [$])")
plt.xticks([1,2,3,4] , ['Q1','Q2','Q3','Q4'])
plt.show()
```
