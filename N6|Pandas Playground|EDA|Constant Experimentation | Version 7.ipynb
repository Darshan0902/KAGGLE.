
# <a id="12"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" > 12 - Groupby().  <br><div> 



````python

# Create a sample DataFrame
data = {
    'Group': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value': [1, 2, 3, 4, 5, 6]
}

# converting the Dictionary into a dataframe
df = pd.DataFrame(data)

# Calculate the mean value for each group and assign it to a new column
df['GroupMean'] = df.groupby('Group')['Value'].transform('mean')

df
````


# <a id="13"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" > 13 - Crosstab().  <br><div> 

````python
# Create a sample DataFrame
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value': [1, 2, 3, 4, 5, 6]
}
# converting the Dictionary into a dataframe
df = pd.DataFrame(data)

# Compute a cross-tabulation table
cross_tab = pd.crosstab(df['Category'], df['Value'])
cross_tab
````

# <a id="14"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" > 14 -  where():  <br><div> 

````python
import numpy as np

# Create a sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [6, 7, 8, 9, 10]
}

# converting the Dictionary into a dataframe
df = pd.DataFrame(data)

# Replace values in column 'B' where the value in column 'A' is less than 3
df['B'] = df['B'].where(df['A'] < 3, np.nan)
df
````


# <a id="15"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" > 15 - nunique()  <br><div> 
This function returns the number of unique elements in each column of a DataFrame. It can be used to quickly get the count of unique values in a column.

````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 2, 3, 3, 3],
                   'B': ['apple', 'banana', 'banana', 'cherry', 'cherry', 'cherry']})

unique_counts = df.nunique()
print(unique_counts)
````

# <a id="16"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" > 16 - Value Counts()  <br><div> 


````python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 2, 3, 3, 3],
                   'B': ['apple', 'banana', 'banana', 'cherry', 'cherry', 'cherry']})

value_counts = df['B'].value_counts()
print(value_counts)
````

# <a id="17"></a>
# <div style= "font-family: Times New Roman; font-weight:bold; letter-spacing: 0px; color:WHITE; font-size:120%; text-align:left;padding:3.0px; background: #FF4500; border-bottom: 8px solid WHITE" > 17 - cut()  <br><div> 

cut(): This function can be used to segment and sort data into bins. It is particularly useful for creating categorical variables based on numerical data.

````python
import pandas as pd

# Create a sample DataFrame
data = {'Age': [21, 25, 29, 33, 37, 41, 45, 49, 53, 57]}
df = pd.DataFrame(data)

# Define the bin edges for age groups
bin_edges = [20, 30, 40, 50, 60]

# Create a new column 'Age Group' by cutting the 'Age' column into bins
df['Age Group'] = pd.cut(df['Age'], bins=bin_edges)

# Display the DataFrame
df
````
